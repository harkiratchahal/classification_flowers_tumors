## Importing the required libraries

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings("ignore")

import pandas as pd
from sklearn.linear_model import LogisticRegression
```

# Flowers

### Load the data in a csv file into a datframe

```python
df = pd.read_csv("iris.csv")
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df.Species.value_counts().plot.bar()
```

### Identify the target column and the data columns

```python
target = df["Species"]
```

```python
features = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
```

### Build and train a classifier

```python
classifier = LogisticRegression()
```

```python
classifier.fit(features, target)
```

### Evaluate the model

```python
classifier.score(features, target)
```

```python
classifier.predict([[5.4,2.6,4.1,1.3]])
```

# Tumor


### Load the dataset

```python
df2 = pd.read_csv("cancer.csv")
```

```python
df2.sample(5)
```

### Identify the target column and the data columns

```python
target = df2["diagnosis"]

features = df2[["radius_mean", "perimeter_mean", "area_mean", "smoothness_mean", "compactness_mean", "concavity_mean", "symmetry_mean"]]
```

```python
classifier_tumor = LogisticRegression()

classifier_tumor.fit( features, target)
```

```python
target.sample(5)
```

```python
features.sample(5)
```

```python
classifier_tumor.predict([[13.45, 86.6, 555.1, 0.1022, 0.08165, 0.03974, 0.1638]])
```

```python
classifier_tumor.score(features, target)
```
